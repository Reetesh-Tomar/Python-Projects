```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
x=pd.read_csv(r'D:\titanic\train.csv')
```

```python
x.head()
```

```python
x.describe()
```

```python
y=x.pop("Survived")
```

```python
y.head()
```

```python
#taking only the numeric values
numeric_var=list(x.dtypes[x.dtypes!="object"].index)     
```

```python
x[numeric_var].head()
```

```python
#filling NaN values with mean
x["Age"].fillna(x.Age.mean())
```

```python
x.describe()
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x[numeric_var],y)
```

```python
z=model.predict(x[numeric_var])
```

```python
print("train Accuracy : ",accuracy_score(y,z))
```

```python
test=pd.read_csv(r'D:\titanic\test.csv')
```

```python
test[numeric_var].head()
```

```python
#filling NaN values with mean
test["Age"].fillna(test.Age.mean())
```

```python
#wherever null values fill with Mean
test=test[numeric_var].fillna(test.mean()).copy()
```

```python
y_pred=model.predict(test[numeric_var])
```

```python
submission=pd.DataFrame({
    "PassengerId":test["PassengerId"],
    "Survived":y_pred
})
submission.to_csv(r'D:\titanic.csv',index=False)
```
